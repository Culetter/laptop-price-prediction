```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/laptop_price.csv', encoding='ISO-8859-1')

pd.set_option('display.max_rows', 1000)

df
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Price_euros'], bins=30, kde=True)
plt.title("Laptop price distribution")
plt.savefig('../results/PriceDistribution.jpg', format='jpg', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Ram', data=df)
plt.title("Ram distribution")
plt.savefig('../results/RamDistribution.jpg', format='jpg', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Company', y='Price_euros', data=df)
plt.title("Price by company")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../results/PriceByCompany.jpg', format='jpg', dpi=300)
plt.show()
```
